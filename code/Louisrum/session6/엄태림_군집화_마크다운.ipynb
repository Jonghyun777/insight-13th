<h1>군집화</h1>

<h2>1. 군집화(Clustering)</h2>

<h3>(1)머신러닝: 비지도학습</h3>
머신러닝: 인공지능의 한 분야로, 컴퓨터가 스스로 학습할 수 있도록 도와주는 알고리즘이나 기술을 개발하는 분야

비지도학습
-
학습하는 데이터에 정답이 주어지는 지도학습과는 달리, 비지도학습은 정답이 없는 데이터를 이용함
-> 데이터 자체에 내재된 구조를 학습하는 방식

특징
- 패턴이 알려지지 않았거나, 계속해서 변화하는 문제를 해결할 때 유용
- 열린 문제를 해결하고 지식을 일반화하는 데 유리함
- 정답 레이블이 없어 모델의 성능을 명확히 측정할 수 없음

<h3>(2)군집화</h3>
군집 : 유사한 데이터들의 집합, 서로 다른 군집에 속한 개체들은 비슷하지 않아야 한다.

군집화 : 데이터를 비슷한 특성을 가진 군집으로 나누는 기법 -> 각 그룹이 어떻게 구성되어있는지 파악하면서 전체 데이터 구조를 이해할 수 있음
- 하나의 관측치가 다른 관측치 또는 그룹의 데이터와 얼마나 유사한지 비교하고, 이 유사성을 기반으로 개체를 함께 그룹화함
- 지도학습에서는 독립변수와 종속변수 관계 사이의 패턴을 분석하지만, 군집화에는 종속변수가 없기 때문에 특정 목표 없이 패턴 인식
- 사전에 정해진 목표 없이 데이터셋에 존재하는 고유한 패턴을 발견하고 추출하는 도구임

군집화의 목표
- 응집도(cohension) 최대화: 같은 군집에 속하는 데이터끼리는 최대한 비슷하도록 함
- 분리도(separation) 최대화ㅣ 서로 다른 군집은 최대한 분리되도록 함

군집화를 통해 데이터 간의 유사성을 최대한 잘 유지하면서, 서로 다른 그룹은 구분될 수 있도록 만드는 것이 목표

<h3>(3)군집화 과정</h3>
군집화의 과정


피처 선택 또는 추출 -> 군집화 알고리즘 선택 -> 군집 유효성 검증 -> 결과 해석
- 

군집화 할 때 꼭 고려하세요!
- 변수 유형 이해: 변수의 종류와 특징을 이해하고, 그에 맞는 방법론을 결정
- 거리 또는 유사도 정의와 측정 : 어떤 방식으로 이걸 정하고, 측정할건지가 중요함(변수의 특성에 관계가 있기 때문)
- 차원 축소 : 유사한 변수들을 묶어서 차원을 축소

군집화는 많은 경우 여러 번의 반복적인 시도를 요구함, 게다가 명확한 기준이 없어 적합한 도구를 고르는 것 조차 하나의 문제임






<h2>2. 군집화 알고리즘</h2>

<h3>(1)계층적 군집화(Hierarchical clustering)</h3>
계층적 군집화 : 데이터 간의 유사성을 기반으로 트리 구조를 형성하며, 상향식 또는 하향식 방식으로 군집을 형성해 나가는 방법

계층적 군집화 알고리즘은 데이터셋의 관측치를 사용해 덴드로그램을 만든다.

덴드로그램이란? : 가까운 두 개체 또는 군집을 점차적으로 병합해가는 과정을 시각적으로 표현한 트리 구조
- 해석 : 수직축 높이가 두 군집이 병합될 때의 군집 간 유사성


계층적 군집화는 군집 개수를 사전에 설정하지 않고, 종료된 후에 원하는 개수를 선택함

트리 구조가 상향식인지, 하향식인지에 따라 구분

상향식이면 - 응집형 계층적 군집화(또는 병합 클러스터링): 각 샘플이 독립적인 클러스터가 되었다가, 하나의 클러스터가 남을 때 까지 가장 가까운 클러스터를 병합함

하향식이면 - 분리형 계층적 군집화(또는 분할 클러스터링): 전체를 하나로 스타트해서, 유사성이 낮은 데이터들을 나눔

응집형 계층적 군집화 : 병합 연산을 통해 모든 데이터포인트를 하나의 그룹으로 합쳐감, 군집 간의 거리를 어떻게 정의하는지에 따라 알고리즘 다름, 6가지 거리 정의 방법을 정리함

1. Single linkage(단일 연결법/최단 연결법) : 서로 다른 군집에서 가장 가까운 두 점 사이의 거리를 측정함. 고립된 군집 찾는데 중점이고, 이상치에 취약함
2. Complete linkage(완전 연결법/최장 연결법) : 각 군집에서 하나의 관측값을 뽑았을 때 나타나는 거리의 최댓값이 기준, 군집들의 내부 응집성에 중점, 이상치에 민감함
3. Average linkage(평균 연결법) : 모든 항목에 대한 거리평균, 계산량이 불필요하게 많아지나, 이상치에 덜 민감함
4. Centroid Method(중심 연결법) : 두 군집의 중심 간의 거리를 측정, 두 군집이 결합되면, 새로운 군집의 평균은 가중평균으로 구함, 군집의 중심을 계산하느라 시간이 오래걸림
5. Median(중앙 연결법) : 샘플의 중앙값을 기준으로 정의, 극단값의 영향을 덜받으나, 기하학적 구조를 파악하긴 어려움
6. Ward's procedure(Ward 연결법) : 군집의 병합 후  군집 내 SSE의 증가분이 최소인 것을 선택, 보통 두 군집이 합해지면 병합된 군집의 SSE는 병합 이전보다 커지는데, 그 증가량이 작아지는 방향으로 형성


완전 연결 방식을 사용한 응집형 계층 군집화의 과정
1. 모든 샘플의 거리 행렬을 계산
2. 모든 데이터 포인트를 단일 클러스터로 표현
3. 가장 비슷하지 않은 샘플 사이 거리에 기초하여 가장 가까운 두 클러스터를 합침
4. 유사도 행렬을 업데이트
5. 하나의 클러스터가 남을 때까지 2~4번 반복


<h3>(2)k-means</h3>
k-means : 데이터를 정해진 개수의 그룹으로 나누되, 각 그룹의 중심점과의 거리가 가장 가까운 데이터끼를 묶어나가는 군집화 기법

주요 단계
1. 데이터 표본들 중에서 랜덤하게 k개의 중심점을 초기 클러스터 중심점으로 선택
2. 각 표본들을 가장 가까운 중심점에 할당(거리는 보통 유클라디안 거리의 제곱을 사용)
3. 각 클러스터에 할당된 표본들의 데이터 평균을 계산하여 중심점을 이동시킴
4. 클러스터 할당이 변하지 않거나, 사용자가 지정한 허용오차 또는 최대 반복횟수에 도달할 때까지 2,3번을 반복

구체적 계산 과정

클러스터 내 SSE를 최소화하는 것을 목표로 중심점을 이동시키며, 각 데이터포인트와 군집 중심의 거리를 최소로 만드는 군집을 찾아냄

적절한 k값은 어떻게 구해요?
- 엘보우 방법과 실루엣 계수를 활용!

k-means 알고리즘을 사용할 때는 특성이 같은 스케일로 측정되었는지 확인해야하고, 필요하면 z-score 표준화나 MinMaxScale로 변환해야함

- 장점 : 직관적, 구현 쉬움, 대용량에 적용 가능, 원형 클러스터를 구분하는데 뛰어남
- 단점 : 사전에 사용자가 클러스터의 개수를 직접 지정해야 함 -> 적절하지 않은 k를 고르면 성능이 좋지 않아 주의해야함, 초기 centroid 값에 민감함, 군집 수 결정이 어려움, 아웃라이어에 민감, 기하학적 모양의 군집 파악 어려빈 클러스터가 존재할 수도 있음(사이킷런 쓰면 해결됨)



k-means++ 알고리즘 : 초기 중심점을 서로 멀리 떨어진 곳에 위치시킴

-> k-means이 가진 단점, 즉 초기 중심점을 랜덤하게 할당해, 초기 중심점에 따라 품질이 달라지는 문제를 해소함

k-means++ 알고리즘의 초기화 과정
1. 선택한 k개의 중심점을 저장할 빈 집합 M을 초기화
2. 입력 샘플에서 첫번째 중심점을 랜덤하게 선택해 M에 할당
3. M에 없는 각 샘플에 대해 M에 있는 중심점까지의 최소제곱 거리를 찾음
4. 가중치가 적용된 확률 분포를 사용해 다음 중심점을 랜덤하게 선택
5. k개의 중심점을 선택할 때까지 3, 4번을 반복
6. 기본 k-means 알고리즘을 수행

엘보우 방법 : 클래스 내 SSE를 바탕으로 그래프를 그려 최적 클러스터 개수를 추정할 수 있음

<h3>(3)DBSCAN</h3>
DBSCAN : 밀도가 높은 지역의 데이터를 하나의 군집으로 묶고, 밀도 기준을 만족하지 못하는 점은 군집에 포함시키지 않는 군집화 알고리즘

주요 단계
1. 어떤 샘플의 특정 반경을 기준으로, 그 안에 있는 이웃 샘플이 지정된 개수 이상이면 핵심 샘플, 지정 개수 이하지만, 다른 핵심 샘플의 반경 안에 있으면 경계 샘플, 둘 다 해당하지 않으면 잡음 샘플로 분류함
2. 개별 핵심 샘플 또는 핵심 샘플의 그룹을 클러스터로 만듬
3. 경계 샘플을 해당 핵심 샘플의 클러스터에 할당

- 장점 : 클러스터 모형을 가정하지 않고, 밀도에 따라 할당해 임의의 기하학적 분포를 갖는 데이터도 잘 처리함, 이상치에 민감하지 않음, 잡음샘플 구분으로 모든 샘플을 클러스터에 할당하지 않음

- 단점 : epsilon과 min_samples 설정에 많은 영향을 받고, 입실론 조절이 어려우며, 적절한 조절을 위해선 데이터셋에 대한 도메인 지식이 필요한 경우가 많음, 연산량이 많아 속도가 느리고, 다른 밀도 분포를 가진 데이터의 군집분석을 잘 못함(밀도가 높은 데이터만 살리고, 나머지를 버려버릴 수 있음), 차원의 저주를 벗어나지 못함(데이터가 고차원으로 이동할수록 데이터포인트 사이의 거리가 늘어나 밀도가 매우 낮아짐

<h3>(4)Gaussian Micture Model(GMM)</h3>
데이터가 여러 다른 모양의 가우시안 분포로 구성되었다고 가정하고, 각 분포를 클러스터로 인식하는 군집화 방법

각 군집을 확률분포로 간주하고, 전체 데이터 분포를 이질적인 여러 개의 확률분포의 혼합으로 보고 모델링함, 데이터의 생성 메커니즘을 모델링할 수 있음

GMM의 가정
- 관측된 데이터포인트들은 특정 가우시안 확률분포에 의해 생성되었다고 가정
- 군집화를 적용하고자 하는 전체 데이터셋에는 여러 개의 가우시안 분포가 섞여있고, 개별 데이터는 우도에 따라 k개의 가우시안 분포 중 한 가지에 속함 -> 섞인 데이터 분포에서 각각의 가우시안 분포를 추출하고, 각각의 분포에 기반해 군집화를 수행함

우도 : 고정된 관측값이 어떠한 확률분포에서 어느 정도의 확률로 나타나는지에 대한 확률, 확률은 고정된 확률분포에서 어떤 관측값이 나타날지의 확률이고, 우도는 값은 고정되어있고, 어떤 확률분포에서 어느 정도의 확률도 나타내는지를 다룸

GMM의 진행 과정
1. 주어진 전체 데이터셋의 분포를 확인
2. 전체 데이터셋은 서로 다른 정규분포 형태의 확률 분포 곡선으로 구성되어있다고 가정
3. 전체 데이터셋을 구성하는 여러 개의 정규분포 곡선을 추출하고, 개별 데이터가 이 중 어떤 정규분포에 속하는지 결정, 이때 각각의 분포가 하나의 군집이 됨

모델의 파라미터 추정

GMM에서는 기댓값-최대화 알고리즘(EM)을 사용하여 모델의 파라미터를 추정함, 이를 통해 각 군집의 모양, 크기, 방향을 모델링 할 수 있음

- 장점 : k-means보다 유연하게 다양한 데이터세트에 잘 적용시킬 수 있음, 타원형 분포나 중첩된 군집 구조에서 좋은 성능
- 단점 : 군집화를 위한 수행시간이 오래 걸림, 가정한 분포에 맞지 않은 데이터면 계산 복잡도가 높아지고, 성능이 저하됨



















<h2>3. 군집화의 평가 방법</h2>
지도학습과는 달리, 성능 평가를 위해 답을 사용할 수 없음

-> 알고리즘 자체의 내부 평가 지표를 사용

정답 레이블이 존재하는 경우에 비지도 학습 방법을 적용한 후, 외부 평가 지표를 활용해 군집화 결과를 평가할 수도 있음


내부 평가
- 군집 내부의 응집도와 군집 간의 분리도를 측정해 데이터 자체의 구조를 기반으로 군집 품질을 평가(실루엣 계수: 한 점이 자기 클러스터 내에서 얼마나 가깝고, 다른 클러스터와는 얼마나 멀리 떨어져 있는지 측정

외부 평가
- 정답 레이블이 있는 경우, 실제 라벨과 군집화 결과를 비교함(ARI)

<h3>(1) 실루엣 계수</h3>
실루엣 계수는 클러스터 내의 샘플들이 얼마나 조밀하게 모여있는지 측정하는 도구로, 이를 통해 군집의 품질을 확인할 수 있음

계산 과정
1. 클러스터 응집력(동일 클러스터 내 다른 포인트 사이 거리의 평균)을 계산
2. 클러스터 분리도(가장 가까운 클러스터의 모든 샘플 간 평균 거리를 계산
3. 클러스터 응집력과 분리도 사이의 차이를 둘 중 큰 값으로 나누어 실루엣을 계산

실루엣 계수는 -1과 1 사이 값을 가짐
- 응집도와 분리도가 같을 경우, 실루엣 계수는 0이 됩니다
- 이상적인 실루엣 계수는 1, 분리도가 응집력에 비해 클 수록 1에 가까워짐

사이킷런을 통해 쉽게 계산 가능

<h3>(2) Dunn Index</h3>
클러스터 간 최소 거리와 최대 거리의 비율을 계산해 품질 평가

값이 클 수록 좋음(분리도가 높고 응집도가 높아야함)

한계 : 군집 수가 많아질수록 계산 비용이 증가하고, 이상치에 민감
