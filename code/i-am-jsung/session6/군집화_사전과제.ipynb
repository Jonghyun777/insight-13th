비지도학습: 정답이 없는 데이터 이용. 데이터 자체에 내재된 구조 학습. 표현 학습 수행함으로서 데이터 고유 패턴 식별  

### 군집화(Clustering)
- 데이터를 비슷한 특성을 가진 그룹으로 나눔
- 유의성 기반으로 그룹화: 하나의 관측치가 다른 관측치, 그룹의 데이터와 얼마나 유사한지 비교
- 종속변수 없어 사전에 정해진 목표 없이 고유한 패턴을 발견하고 추출 -> 비지도 학습
- 목표
  - 응집도 최대화: 같은 군집 최대한 비슷
  - 분리도 최대화: 다른 군집 최대한 분리

### 군집화 과정: 피처 선택 or 추출 -> 군집화 알고리즘 선택 -> 군집 유효성 검증 -> 결과 해석
** 고려사항  
- 변수가 연속형 or 명목형인지. 개수와 특징 -> 방법론 결정
- 거리(유사도) 정의 및 측정
- 유사한 변수들을 묶어 차원 축소

=> one-shot X, 여러번의 반복적인 시도 요구

## 군집화 알고리즘
### 1) 계층적 군집화
- 데이터 간의 유사성을 기반으로 트리 구조를 형성. 상향 or 하향식 방식으로 군집 형성
- 데이터 관측치로 Dendrogram을 만듦 ~ 가까운 군집을 점차적으로 병합해 나감

응집형 계층적 군집화(병합 클러스터링): 상향식으로 구성
    - 각 샘플들 = 독립적인 클러스터
    - 하나의 클러스터(데이터 전체)가 남을 때까지 가장 가까운 클러스터 병합
    
분리형 계층적 군집화(분할 클러스터링): 하향식 구성
    - 전체 샘플을 포함하는 하나의 클로스터 -> 유사성이 낮은 데이터를 작은 클러스터로 나눔

### 응집형 계층적 군집화(상향식, 병합 클러스터링)
- N개 클러스터 -> 1개 데이터 전체 (병합 연산) 
- "군집 간의 거리를 어떻게 정의" (알고리즘)
      - 단일 연결 알고리즘 / 완전 연결 알고리즘
  
1) 단일 연결 알고리즘: 클러스터 쌍에서 가장 비슷한 샘플 간 거리 계산 이후 거리가 가장 작은 두 클러스터 병합
  - 이상치에 취약
    
2) 완전 연결 알고리즘: 가장 비슷하지 않은 샘플을 비교해 병합 수행, 이상치에 민감
    - 모든 샘플의 거리 행렬 계산
    - 모든 데이터 포인트를 단일 클러스터로 표현
    - 가장 비슷하지 않은 샘플(거리가 먼...) 사이의 거리 중 가장 가까운 클러스터 -> 병합
    - 유사도 행렬 업데이트
    - 하나의 클러스터가 남을 때까지 반복
      
3) 평균 연결법: 모든 항목에 대한 거리 평균 ~ 계산량 많아짐

4) 중심 연결법: 두 군집의 중심 간 거리 ~ 시간 오래 걸림

5) 중앙 연결법: 모든 샘플의 중앙값으로 군집간 거리 정의 ~ 극단값 영향 X, 기하학적 구조 파악 어렵

6) Ward 연결법: 군집의 병합 후 SSE(오차제곱합)의 증가분이 최소인 것을 선택. 

### 2) K-means
- 데이터를 정해진 K개의 그룹으로 나누고 그룹의 중심점 선택 후 중심에 가장 가까운 데이터끼리 묶음
- 보통 유클리디안 거리의 제곱으로 샘플 간 거리 측정
- 클러스터 내 SSE(제곱 오차합)을 최소화하는 것을 목표로 중심점을 이동시키며 최소화
  
단계
1. 랜덤하게 k개의 중심점을 초기 클러스터 중심으로 선택
2. 각 표본들을 가장 가까운 중심에 할당
3. 각 클러스터에 할당된 표본들로부터 중심점 이동시킴
4. 클러스터 할당이 변하지 않거나 허용 오차, 최대 반복 횟수에 도달할 때까지 반복

장점
- 직관적, 구현이 쉽다
- 대용량 데이터에도 적용 가능

단점
- 초기 centroid 값에 민감
- 군집 수(K) 결정 어려움
- 아웃라이어에 민감 ~ 평균 centroid를 중심으로 구성하기 때문
- 기하학적인 모양의 군집은 파악이 힘듦 -> DBSCAN, K-MEDIAN, K-Medoids
- 비어있는 클러스터가 존재할 수 있다. ~ sklearn에서는 이 문제 해결 

단점 해결
- sklearn의 k-means++
- 엘보우 방법: 그래프를 활용해 최적 클러스터 개수 k 추정 ~ k에 따른 군집 품질
  - k가 증가하면 왜곡 줄어듦.
  - 왜곡이 빠르게 감소하는 지점의 k를 찾음.

### 3) DBSCAN
- 밀도가 높은 지역의 데이터를 하나의 군집으로 묶고, 밀도 기준을 만족하지 못하면 군집에 포함 X
- 데이터의 밀도 분포 기반

단계
1. 조건에 따라 샘플을 핵심 / 경계 / 잡음 샘플 중 하나에 할당
  - 핵심 샘플: 어떤 샘플의 특정 반경 안에 있는 이웃 샘플이 지정된 개수 이상이면 핵심 샘플
  - 경계 샘플: 특정 반경 내 이웃이 지정 개수보다 적지만 다른 샘플의 반경 안에 있으면 경계 샘플
  - 잡음 샘플: 위 두개에 해당 X

2. 개별 핵심 샘플 또는 핵심 샘플의 그룹을 클러스터로 만듦
3. 경계 샘플을 해당 핵심 샘플의 클러스터에 할당

장점
- 클러스터 모양을 원형으로 가정 X, 밀도에 따라 클러스터 할당 -> 기하학적 분포 데이터 처리 가능
- 이상치에 민감 X ~ 잡음 샘플 구분
- 모든 샘플을 클러스터에 할당 X

단점
- 하이퍼 파라미터 epsilon과 min_samples 설정에 영향을 많이 받음
- epsilon을 조절하기 어렵. 데이터셋에 대한 도메인 지식 필요한 경우 많음
- 연산량 많음
- 다른 밀도를 가진 데이터의 군집 분석 잘 못함...(??)
  - 밀도가 높은 곳에 집중하다 보면 밀도가 낮은 곳을 군집으로 인식 못하고 noise point로 구분해버림
- 차원의 저주(유클리디안 거리 알고리즘이 갖는 문제)
  - 같은 양의 데이터가 고차원으로 이동할 수록 거리가 늘어나 밀도가 매우 낮아짐


### 4) Gaussian Mixture Model
- 데이터가 여러 가우시안 분포로 구성되었다고 가정. 각 분포를 클러스터로 인식
- 각 군집을 확률 분포로 간주
- 데이터 생성 매커니즘을 모델링할 수 있다.

가정
- 관측된 데이터 포인트들은 특정 가우시안 확률 분포에 의해 생성되었다.
- 개별 데이터는 우도(가능도)에 따라 K개의 가우시안 분포 중 하나에 속한다.
- 우도(likelyhood): 고정된 관측값이 어떤 확률분포에서 어느정도의 확률로 나타나는지에 대한 확률

단계
- 전체 데이터 셋 분포 확인 / GMM 가정(가우시안 확률분포들로 구성 가정)
- 여러 개의 정규분포 곡선 추출하여 개별 데이터가 어떤 분포에 속하는지 결정

모델의 파라미터 추정
- 기댓값-최대화 알고리즘 사용
- 각 군집의 중심, 모양, 크기, 방향까지 모델링

장점
- 타원형 분포, 중첩된 군집 구조, 길쭉한 분포에서도 적용 가능

단점
- 수행시간 길다.
- 가우시안 분포에 맞지 않은 데이터일 경우 성능 저하

## 군집화의 평가방법
내부 평가 지표(Internal Evaluation) - SSE, 군집내 응집도, 분리도를 정량적으로 측정

외부 평가 지표: 데이터셋에 정답이 존재하나 비지도 학습 방법 적용 (ARI, Adjusted Rand Index)

1) 실루엣 계수: 클러스터 내 샘플들이 얼마나 조밀하게 모여있는지 측정 -1 ~ 1 사이의 값.
    - 응집력: 샘플 x와 동일한 클러스터 내 모든 포인트 사이의 거리 평균 계산
    - 분리도: 가장 가까운 클러스터의 모든 샘플 간 평균 거리 계산
    - 응집력과 분리도 차이(분리도 - 응집력)를 MAX(응집력, 분리도)로 나누어 실루엣을 계싼함.
    - 분리도 >> 응집력, 1에 가까워짐

2) Dunn Index: 클러스터 간 최소거리(분리도)와 최대거리(응집도)의 비율을 계산해 클러스트링의 품질 평가
    - 클수록 좋은 군집화 결과
    - 한계: 군집 수가 많을수록 계산 비용 증가 / 이상치 민감
