## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 군집화란? </span>

군집화 (clustering) : 데이터를 비슷한 특성을 가진 그룹으로 나누는 <U>"비지도 학습 기법"</U>

<details>
  <summary><h4>비지도학습이란?<BR>(☝︎click!)</h4></summary>
  데이터 자체에 내재된 구조를 학습하는 머신러닝 방법
  
    정답 레이블이 없음 -> 모델 성능을 정확하게 측정할 수 없음

    표현 학습(representation learning)을 수행하며 데이터셋의 고유 패턴을 식별함    
</details>




유사성을 기반으로 개체를 함께 grouping하는 방법

#### 군집화의 목표

        응집도 최대화 : 같은 군집에 속하는 데이터끼리 최대한 비슷하게 하기
        분리도 최대화 : 서로 다른 군집은 최대한 분리하기

⇨ 최종적으로, <U>데이터 간의 유사성을 최대한 유지하며</U>, <U>서로 다른 그룹은 구분하기</U>

#### 군집화의 기본 과정

1. 피쳐 선택 or 추출

2. 군집화 알고리즘 선택

3. 군집 유효성 검증

4. 결과 해석

📍군집화에서 중요한 내용들

> 변수 유형 이해 - 종류(연속/명목), 개수, 특징 등으로 <U>데이터에 대해 파악하여 방법론을 결정해야함</U>

> 거리 (유사도) 정의 & 측정 - 데이터 간 유사도 기반으로 군집화를 하기 때문에, <U>어떤 방식으로 거리 or 유사도를 측정하는지 중요함!</U>

> 차원 축소 - <U>유사한 데이터들을 묶어</U>서 차원 축소하기 (ex. 연봉, 나이, 연차, 직군... -> 나이 & 연차를 묶어서 차원 축소하기)

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 군집화 알고리즘 </span>

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 계층적 군집화(Hierarchical Clustering) </span>





#### Dendrogram이란?

가까운 <U>2개의 개체/군집의 병합 과정</U>을 트리 구조로 표현한 것

Dendrogram의 높이 = 두 개체/군집의 병합 과정에서의 <U>거리(유사성)</U><BR>
_(위 사진에서는 E, F의 연결 링크 부분의 높이가 가장 짧음 -> E & F가 가장 가깝다는 것을 알 수 있음)_



데이터 간의 유사성을 기반으로 트리 구조(dendrogram) 형성 -> 상/하향식으로 군집 형성하기!!

* 응집형 계층적 군집화 (병합 클러스터링) : 먼저 각 샘플이 독립적인 클러스터가 되고, 하나의 클러스터만 남을 때까지 가장 가까운 클러스터를 병합

* 분리형 계층적 군집화 (분할 클러스터링) : 전체 샘플을 포함하는 하나의 클러스터에서 시작해서 유사성이 낮은 데이터들을 더 작은 클러스터로 분할

응집형 계층적 군집화는 각각 1개의 데이터포인트만을 포함하는 N개의 클러스터에서 시작하는데, 이후 수 회의 병합 연산을 통해 이 클러스터들을 합쳐 더 큰 클러스터로 만든다.

계속해서 병합 연산을 한다면 최종적으로는 모든 데이터포인트들이 같은 그룹에 속하게 될 것인데, <BR>
이 때, 가까운 군집 간 병합이 일어나기 때문에 **군집 간의 거리를 어떻게 정의하느냐**에 따라 알고리즘이 달라진다.

정의 방법에는 대표적으로 6가지가 있다.

* 단일 연결법 : 가장 가까운 두 점 사이의 거리를 군집 간의 거리로 측정하는 방법



* 완전 연결법 : 군집에서 하나씩 관측값을 뽑았을 때 나타날 수 있는 거리의 최댓값으로 측정하여 가장 유사성이 큰 군집으로 병합해 나가는 방법



* 평균 연결법 : 모든 항목에 대한 거리 평균



* 중심 연결법 : 두 군집의 중심 간의 거리를 측정



* 중앙 연결법 : 군집간의 거리를 군집 내 모든 샘플의 중앙값으로 정의

* 와드 연결법 : 군집의 병합 후 군집 내 SSE(오차제곱합)의 증가분 최소인 것을 선택



### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> K-means </span>

데이터를 정해진 개수(K)의 그룹으로 나누되, <U>각 그룹의 중심점(centroid)과의 거리가 가장 가까운 데이터끼리 묶는</U> 군집화 알고리즘

k-means 알고리즘 주요 단계



- **k-means의 장점**
    - 직관적이고 구현이 쉬움.
    - 대용량 데이터에도 적용이 가능.

- **k-means의 단점**
    - **초기 centroid 값에 민감함** : 초기 centroid를 설정하는 많은 방법이 존재, 그에 따라 형성되는 최종적인 군집이 달라질 수 있음.
    - **군집 수 (k) 결정이 어려움** : 사용자가 직접 k값을 지정해야 하며, 이 k값이 결과의 질에 큰 영향을 미침.
    - **아웃라이어에 민감** : 평균 중심으로 군집을 구성하기 때문에, **아웃라이어 하나가 중심을 왜곡시킬 수 있음**.
    - **기하학적인 모양의 군집은 파악하기 어려움.**

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> DBSCAN </span>

밀도가 높은 지역의 데이터를 하나로 묶고, 기준 밀도에 못 미치는 점은 군집화에서 제외하는 방식의 군집화 알고리즘

DBSCAN의 장점
* 클러스터 모양을 원형으로 가정하지 않아서, 임의의 기하학적 분포를 갖는 데이터도 잘 처리함
* 이상치에 민감하지 않음
* 모든 샘플을 클러스터에 할당하지 않음

DBSCAN의 단점
* epsilon과 min_samples 설정에 민감함
* epsilon을 조절하기 쉽지 않고, 조절을 위해서는 도메인 지식이 필수적임
* 연산량이 많아 속도가 느림
* 차원의 저주 문제

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> Gaussian Mixture Model (GMM) </span>

데이터가 다양한 모양의 가우시안 분포로 구성되었다고 가정, 각 분포를 클러스터로 인식하는 군집화 알고리즘

GMM의 진행 과정



1. 주어진 전체 데이터셋의 분포 확인



2. 전체 데이터셋은 서로 다른 정규 분포 형태의 확률 분포 곡선으로 구성됐다고 가정



3. 전체 데이터셋을 구성하는 여러 개의 정규분포 곡선 추출, 개별 데이터가 이 중 어떤 정규분포에 속하는지 결정한다. 이때 각각의 분포가 하나의 군집이 됨.

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 군집화 평가 방법 </span>

비지도 학습에는 별다른 정답 데이터셋이 없기 대문에, 다양한 평가 지표가 없다.<BR>
그렇기 때문에, 군집화 과정에서 군집 품질 평가를 위해서는 <U>알고리즘 자체의 내부 평가 지표를 사용해야한다.</U><BR>
내부 평가 지표에는 실루엣 계수, 던 지수(Dunn Index)가 있다.


데이터셋의 정답 레이블이 존재하는 경우에도 비지도 학습 방법을 적용할 수 있다!!<BR>
-> 이럴땐 외부 평가 지표를 활용할 수 있음

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> 실루엣 계수 </span>

클러스터 내 샘플들이 얼마나 조밀하게 모여 있는지를 측정하는 도구입니다. 이를 통해 군집의 품질을 확인할 수 있음

#### 군집화가 잘 된 경우



실루엣 계수의 값 분포가 균형적이고, 평균 실루엣 계수가 0.7 이상이다.

#### 군집화가 잘 안 된 경우



실루엣 계수의 분포가 불균형적이고, 1번 클러스터는 많은 데이터 포인트들이 값이 낮음

### <span style="background-color:#FFFFE0; color:#CD5C5C; padding:4px; border-radius:5px;"> Dunn Index(던 지수) </span>

클러스터 간 최소 거리(분리도)와 클러스터 내 최대 거리(응집도)의 비율을 계산해 클러스터링의 품질을 평가하는 지표

${Dunn Index} = \frac{
\min\limits_{i \ne j} \; d(C_i, C_j)}{\max\limits_{1 \le k \le K} \; \delta(C_k)}$

Dunn Index 값이 클수록 좋은 군집화 결과라는 것을 의미한다.<BR>
(군집 간의 거리가 크고, 군집 내부의 거리가 작을수록 좋음)

#### Dunn Index의 한계
* 군집 수가 많아지면, 계산 비용이 증가
* 이상치에 민감할 수 있음
